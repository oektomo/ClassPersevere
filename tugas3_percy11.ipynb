### Data Science
>> Contoh produk yang diimplementasikan pada bidang data science yaitu penggunaan e-commerce seperti Shopee, Tokopedia, Bukalapak dan lan sebagainya. Aplikasi E-commerce ini memanfaatkan data hasil penggunaan user untuk kemudian digunakan pada rekomendasi produk yang sesuai diinginkan pengguna.
>> Contoh lainnya adalah google analytics yang dapat menyediakan data berdasarkan hasil kunjungan anda dalam website. 

### Computer Vision
>> Pada Computer Vision, produk yang dihasilkan yaitu seperti mobil yang dapat mengendarai sendiri atau self-driving cars. Contoh lainnya yaitu tilang online, dimana dengan menggunakan kamera cctv dapat mendeteksi pelanggaran yang ada di lalu lintas.

### NLP (Natural Language Processing)
>> Contoh dari NLP yaitu smart assistan seperti Siri, Alexa, Google Assitant, dan lain sebagainya yang bisa mendengarkan suara masukan dan mengenali bahasa manusia.
>> Contoh lainnya yaitu fitur Autocomplete pada saat pengetikan di smartphone.
